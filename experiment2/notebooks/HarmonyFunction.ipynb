```python
import sys
sys.path.append('../')
import numpy as np
import matplotlib.pyplot as plt

from planning_alg3.planning import Planning

planning = Planning(None, None)

targets = [0.1, 1, 5, 10, 100]
output_diff_percents_range = [0.1, 10]

percents = np.linspace(output_diff_percents_range[0], output_diff_percents_range[1], 100)
netoutputs = {}
harmonies = {}
derivative_harmonies = {}
for target in targets:
    netoutputs[target] = percents * target
    harmony_for_target = lambda x: planning.harmony(target, x)
    harmonies[target] = np.array(list(map(harmony_for_target, netoutputs[target])))
    derivative_harmony_for_target = lambda x: planning.derivative_harmony(target, x)
    derivative_harmonies[target] = np.array(list(map(derivative_harmony_for_target, netoutputs[target])))

    plt.axvline(x=target)
    plt.plot(netoutputs[target], harmonies[target])
    plt.show()
    
    plt.axvline(x=target)
    plt.plot(netoutputs[target], derivative_harmonies[target])
    plt.show()
```

## Conclusions:
- The value of the derivative depends on the scale of the target.
- For small deviations from the target the difference between positive and negative might not be enough
